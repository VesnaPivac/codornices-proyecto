### Time

```python
def convert_to_time_format(time_value):
    # Convertir a string y eliminar cualquier parte decimal
    time_str = str(int(time_value))
    
    # Asegurarse de que el string tenga al menos 4 caracteres (llenar con ceros a la izquierda si es necesario)
    time_str = time_str.zfill(4)
    
    # Extraer horas y minutos
    hours = time_str[:2]
    minutes = time_str[2:]
    
    # Formatear como HH:MM
    return f"{hours}:{minutes}"
```

```python
# Aplicar la funciÃ³n a toda la columna 'time'
df_1_clean['time'] = df_1_clean['time'].apply(convert_to_time_format)

# Verificar los cambios
df_1_clean.head()
```
